```python
import pandas as pd
from random import shuffle, choice
import numpy as np
```

<H1> Data_Processing


```python
df = pd.read_excel('/content/datahkn.xlsx',sheet_name="Final", index_col= False)
```

```python
df['Thể loại món ăn'].unique()
```

```python
df['Thể loại món ăn'].replace(['Chè', 'Giải khát-Tráng miệng'], ['Tráng miệng', 'Tráng miệng'], inplace=True)
```

```python
df['Thể loại món ăn'].replace(['Canh', 'Lẩu'], ['Canh-Lẩu', 'Canh-Lẩu'], inplace=True)
```

```python
df
```

```python
data = {

}
for i, row in df.iterrows():
    if row['Thể loại món ăn'] not in data:
        data[row['Thể loại món ăn']] = []
    data[row['Thể loại món ăn']].append(row['ID món ăn'])
```

```python
def removekey(d, key):
    r = dict(d)
    del r[key]
    return r
```

```python
import numpy as np
result_all = []

for key1, values1 in data.items():
    if key1 is not np.nan:
        shuffle(values1)
        for x in values1:
            for i in range(0, 10):
                data_2 = removekey(data, key1)
                for key2, values2 in data_2.items():
                    y_1 = choice(values2)
                    
                data_3 = removekey(data_2, key2)
                if key1 == 'Canh-Lẩu' or key2 == 'Canh-Lẩu':
                    for key3, values3 in data_3.items():
                        y_2 = choice(values3)
                else:
                    key3 = 'Canh-Lẩu'
                    y_2 = choice(data_3['Canh-Lẩu'])

                data_4 = removekey(data_3, key3)
                if key1 == 'Tráng miệng' or key2 == 'Tráng miệng' or key3 == 'Tráng miệng':
                    for key4, values4 in data_4.items():
                        y_3 = choice(values2)
                else:
                    key4 = 'Tráng miệng'
                    y_3 = choice(data_3['Tráng miệng'])

                result = {
                        'x': x,
                        'y_1': y_1,
                        'y_2': y_2,
                        'y_3': y_3,
                    }
                result_all.append(result) 
```

```python
df1 = pd.DataFrame(result_all)
df1.to_csv('/content/data_use_train.csv', index = False)
```

<H1> Chia train test

```python
features = ['y_1', 'y_2', 'y_3']
X = df2['x']
y = df2[features]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train = np.array(X_train).reshape(X_train.shape[0], 1)

y_train = np.array(y_train)
```

<H1> KNN

```python
knn = KNeighborsClassifier(n_neighbors= 7)

knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(np.array(X_test).reshape(X_test.shape[0], 1))
```

```python
df_pred = pd.DataFrame(y_pred, columns=['y_1', 'y_2', 'y_3'])
```

```python
import tqdm
len_row = len(df_pred)
count_true = 0
columns = ['y_1', 'y_2', 'y_3', 'x']

for i, row_1 in tqdm.tqdm(df_pred.iterrows()):
    type_tmp = set()
    for j, row_2 in df.iterrows():
        for feature in columns:
            if row_2['ID món ăn'] == row_1[feature]:
                type_tmp.add(row_2['Thể loại món ăn'])
    
    if (len(type_tmp) == 4) and ('Canh-Lẩu' in type_tmp) and ('Tráng miệng' in type_tmp):
        count_true += 1

print("Độ chính xác của mô hình là: ", round((count_true/len_row) * 100, 2))
```

```python
def predict(input: str):
    id_mon_an = []
    result_all = []

    for i, row in df.iterrows():
        if (row['Tên món ăn'] == input) or (row['Tên món ăn'] == ' ' + input):
            id_mon_an.append(row['ID món ăn'])
            break
    if len(id_mon_an) == 0:
        print('Không tìm thấy món ăn')
        return
    # dự đoán
    y_pred = knn.predict(np.array(id_mon_an).reshape(np.array(id_mon_an).shape[0], 1))
    for id in y_pred[0]:
        for i, row in df.iterrows():
            if row['ID món ăn'] == id:
                result = {
                    'ID món ăn': id,
                    'Tên món ăn': row['Tên món ăn'],
                    'Thể loại món ăn': row['Thể loại món ăn'],
                    'Giá ước tính': int(row['Giá ước tính']),
                }
                result_all.append(result)
    df_result = pd.DataFrame(result_all)
    total_prices = df_result['Giá ước tính'].sum()
    print("Menu đề xuất là: ")
    print(df_result)
    print(f"Tổng số tiền ước tính cho thực đơn của quý khách là: {total_prices} vnđ")
    # return total_prices
```
